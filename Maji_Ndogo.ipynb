### Task 5: Investigate pollution issues:

```python
%%sql

SELECT *
FROM well_pollution
LIMIT 5;
```

```python
%%sql

SELECT DISTINCT(description) as 'Water contamination classification'
FROM well_pollution
LIMIT 5;
```

Identify records that mistakenly have the word Clean in the description and biological value is greater than 0.01

```python
%%sql

SELECT *
FROM well_pollution
WHERE description LIKE 'Clean%' AND biological > 0.01
```

The second issue we need to fix is in our results column. We need to update the reults column from Clean to Contaminated: Biological where the biological column has a greater value than 0.01

```python
%%sql

SELECT *
FROM well_pollution
WHERE results = 'Clean' AND biological > 0.01
```

```python
%%sql

CREATE TABLE md_water_services.well_pollution_copy AS(
    SELECT *
    FROM md_water_services.well_pollution
)
```

```python
%%sql

SHOW TABLES;
```

```python
%%sql

SELECT *
FROM well_pollution_copy
LIMIT 5;
```

```python
%%sql

UPDATE well_pollution_copy
SET description = 'Bacteria: E. coli' WHERE description = 'Clean Bacteria: E. coli';

UPDATE well_pollution_copy
SET description = 'Bacteria: Giardia Lamblia' WHERE description = 'Clean Bacteria: Giardia Lamblia';

UPDATE well_pollution_copy SET results = 'Contaminated: Biological' WHERE biological > 0.01 AND results = 'Clean';
```

```python
%%sql

-- Test fixes (Should return 0 rows if successful)
SELECT *
FROM well_pollution_copy
WHERE description LIKE 'Clean %' OR (results = 'Clean' AND biological > 0.01);

DROP TABLE md_water_services.well_population_copy;
```
